```typescript
// Preprocess data
const exampleRawData = `
123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  
`.slice(1, -1);

const rawData = await Deno.readTextFile("day_6.txt");

function preprocess(rawData: string) {
  const grid = rawData.split("\n")
      .map(row => [...row.match(/\S+/g)!]);
  return grid;
}
```

```typescript
function part1(data: ReturnType<typeof preprocess>) {
  const ops = data.pop()!;
  const cols: number[] = data[0].map(Number);
  for (let y = 1; y < data.length; y++) {
    const row = data[y];
    for (let x = 0; x < row.length; x++) {
      const op = ops[x];
      if (op === "+") {
        cols[x] += +row[x]
      } else {
        cols[x] *= +row[x];
      }
    }
  }

  let total = 0;
  for (let x = 0; x < cols.length; x++) {
    total += cols[x];
  }
  return total;
}

part1(
  preprocess(
    // exampleRawData,
    rawData,
  )
)
```

```typescript
// Preprocess data
function part2(rawData: string) {
  const grid = rawData.split("\n").map(row => row);
  const ops = grid.pop()!;
  let op = ops[0];
  let sum = 0;
  let agg: number | undefined;
  for (let x = 0; x < ops.length; x++) {
    if (ops[x] !== " ") {
      op = ops[x];
      sum += agg ?? 0;
      agg = undefined
      console.log();
    }
    const raw = grid.map(row => row[x]).join("").trim();
    if (!raw) continue;
    const num = +raw;
    // console.log(num);
    if (agg === undefined) {
      agg = num
    } else if (op === "+") {
      agg += num;
    } else {
      agg *= num;
    }
    // console.log("agg", agg);
  }
  sum += agg!;
  return sum;
}

part2(
  // exampleRawData,
  rawData,
)
```
