```typescript
// Preprocess data
const exampleRawData = `
..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@.
`.slice(1, -1);

const rawData = await Deno.readTextFile("day_4.txt");

function preprocess(rawData: string) {
  return rawData
    .split("\n")
    .map((e) => e.split(""));
}
```

```typescript
function part1(data: ReturnType<typeof preprocess>) {
  let rolls = 0;
  for (let y = 0; y < data.length; y++) {
    for (let x = 0; x < data[0].length; x++) {
      if (data[y][x] !== "@") continue;

      const count = 
        (data[y - 1]?.[x - 1] === "@" ? 1 : 0) +
        (data[y - 1]?.[x] === "@" ? 1 : 0) +
        (data[y - 1]?.[x + 1] === "@" ? 1 : 0) +
        (data[y][x - 1] === "@" ? 1 : 0) +
        (data[y][x + 1] === "@" ? 1 : 0) +
        (data[y + 1]?.[x - 1] === "@" ? 1 : 0) +
        (data[y + 1]?.[x] === "@" ? 1 : 0) +
        (data[y + 1]?.[x + 1] === "@" ? 1 : 0);
      
      rolls += count < 4 ? 1 : 0;
    }
  }
  return rolls;
}

part1(
  preprocess(
    // exampleRawData,
    rawData,
  )
)
```

```typescript
function part2(data: ReturnType<typeof preprocess>) {
  function removeRolls(data: ReturnType<typeof preprocess>) {
    let rolls = 0;
    for (let y = 0; y < data.length; y++) {
      for (let x = 0; x < data[0].length; x++) {
        if (data[y][x] !== "@") continue;

        const count = 
          (data[y - 1]?.[x - 1] === "@" ? 1 : 0) +
          (data[y - 1]?.[x] === "@" ? 1 : 0) +
          (data[y - 1]?.[x + 1] === "@" ? 1 : 0) +
          (data[y][x - 1] === "@" ? 1 : 0) +
          (data[y][x + 1] === "@" ? 1 : 0) +
          (data[y + 1]?.[x - 1] === "@" ? 1 : 0) +
          (data[y + 1]?.[x] === "@" ? 1 : 0) +
          (data[y + 1]?.[x + 1] === "@" ? 1 : 0);
        
        if (count < 4) {
          rolls += 1;
          data[y][x] = "."
        }
      }
    }
    return rolls;
  }
  let rolls = 0;
  while (true) {
    const removed = removeRolls(data);
    if (removed === 0) break;

    rolls += removed;
  }
  return rolls;
}

part2(
  preprocess(
    // exampleRawData,
    rawData,
  )
)
```
