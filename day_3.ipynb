```typescript
// Helper functions
const log = <T extends unknown[]>(...data: T) => (console.log(...data), data);
const mod = (a: number, b: number) => ((a % b) + b) % b;
```

```typescript
// Preprocess data
const exampleRawData = `
987654321111111
811111111111119
234234234234278
818181911112111
`.slice(1, -1);

const rawData = await Deno.readTextFile("day_3.txt");

function preprocess(rawData: string) {
  return rawData
    .split("\n")
    .map((e) => e.split("").map(Number));
}
```

```typescript
function part1(data: ReturnType<typeof preprocess>) {
  let total = 0;
  for (const line of data) {
    let a = line[0];
    let max = 0;
    for (let i = 1; i < line.length; i++) {
      const n = line[i];
      max = Math.max(max, a * 10 + n);
      a = Math.max(a, n);
    }
    total += max;
    // log(max)
  }
  return total
}

part1(
  preprocess(
    // exampleRawData,
    rawData,
  )
)
```
