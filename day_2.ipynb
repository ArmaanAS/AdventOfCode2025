```typescript
// Preprocess data
const exampleRawData = `
11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124
`.slice(1, -1);

const rawData = await Deno.readTextFile("day_2.txt");

function preprocess(rawData: string) {
  return rawData
    .split(",")
    .map((e) => e.split("-").map(Number) as [number, number]);
}
```

```typescript
function part1(data: ReturnType<typeof preprocess>) {
  function isInvalid(i: number) {
    const s = `${i}`;
    if (s.length % 2 === 1) return false;

    return s.slice(0, s.length / 2) === s.slice(s.length / 2);
  }

  let invalid = 0;
  for (const [a, b] of data) {
    for (let i = a; i <= b; i++) {
      if (isInvalid(i)) invalid += i;
    }
  }
  return invalid;
}

part1(
  preprocess(
    // exampleRawData,
    rawData,
  )
)
```

```typescript
function part2(data: ReturnType<typeof preprocess>) {
  function isInvalid(i: number) {
    const s = `${i}`;

    for (let r = 1; r <= s.length / 2; r++) {
      if (s.length % r !== 0) continue;

      const part = s.slice(0, r);
      const invalid = part.repeat(s.length / r) === s;
      if (invalid) return true;
    }
    return false;
  }

  let invalid = 0;
  for (const [a, b] of data) {
    for (let i = a; i <= b; i++) {
      if (isInvalid(i)) invalid += i;
    }
  }
  return invalid;
}

part2(
  preprocess(
    // exampleRawData,
    rawData,
  )
)
```
